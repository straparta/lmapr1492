#### Déterminer les 3 premiers pics du diffractogramme correspondant (en indiquant les indices hkl associés) pour un rayonnement X incident Cu Kα de longueur d’onde λ=1.54060 Å.

```python
from tabulate import tabulate
import matplotlib.pyplot as plt
from pymatgen.analysis.diffraction.xrd import *

xrd = XRDCalculator(symprec=0.1)
xrdpattern = xrd.get_pattern(structure)
hkls = xrdpattern.hkls
```

Nous allons ici étudier le spectre de diffraction infrarouge du cristal. Ce spectre est obtenu grâce à la figure d'une longueur d'onde λ=1.54060 A˚.
Pour certains angles incidents 2θ, la lumière incidente interfère à cause des différents plans cristallins (spéarés d'une distance d). Les interférences constructives sont ajoutées sous forme de raies au spectre d'interférence.


On observe que certaines raies du spectre sont beaucoup plus faibles que celles adjacentes. On peut considérer qu'elles sont probablement due à des interférences parasites.

On peut définir les indices (h,k,l) pour les différents pics

```python
for i in range(0,3):
        print(tabulate([
            [" ",
             str(xrdpattern.x[i]),
             str(xrdpattern.y[i]),
             str(xrdpattern.hkls[i][0]['hkl']),
             str(xrdpattern.hkls[i][0]['multiplicity']),
             xrdpattern.d_hkls[i]]],
        headers=["Pic " + str(i+1),"2θ","Intensité","Indice (h,k,l)","Multiplicité", "Distance Interplanaire"]))
        print("\n")
        
xrd.show_plot(structure,two_theta_range=[0,45])

# Pour plus de détails
xrd.show_plot(structure)
```
