#### Déterminer  les positions des 3 atomes dans la maille conventionnelle

```python
Ti = (0.5,0.5,0.5)
Fe = (0.25,0.25,0.25)
Te = (0,0,0)
```

#### Déterminer  les opérations de symétrie

A partir du fichier CIF et de JSmol, on peut determiner 3 opérations de symétrie (hors identité). 

```python
S1 = SymmOp.from_xyz_string('y, -x, -z')
S2 = SymmOp.from_xyz_string('-x, -y, z')
S3 = SymmOp.from_xyz_string('-y, x, -z')
```

#### Appliquer  les opérations de symétrie

On applique les 3 symétries chacun pour 1 atome différent

.   Symétrie 1 sur atome de Ti

.   Symétrie 2 sur atome de Fe 

.   Symétrie 3 sur atome de Te 


```python
Ti_S1 = S1.operate(Ti)
Fe_S2 = S2.operate(Fe)
Te_S3 = S3.operate(Te)
```

```python
print(tabulate([
	["S1 : ("+str(S1.as_xyz_string()+")"),"Ti", str(Ti),str(Ti_S1)],
	[S1,"",""],
	["----------------------------------","-------","-----------------","------------------"],
	["S2 : ("+str(S2.as_xyz_string()+")"),"Fe", str(Fe),str(Fe_S2)],
	[S2,"",""],
	["----------------------------------","-------","-----------------","------------------"],
	["S3 : ("+str(S3.as_xyz_string()+")"),"Te", str(Te),str(Te_S3)],
	[S3,"",""],
	["----------------------------------","-------","-----------------","------------------"]],
	headers=["Opération de symétrie","Atome","Initial","Symétrisé"]))
print("\n")
```
